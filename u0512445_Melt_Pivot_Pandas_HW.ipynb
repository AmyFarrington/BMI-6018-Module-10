```python
#Import required libraries
import pandas as pd
```

```python
#Import required files
cp_df = pd.read_csv('hungary_chickenpox.csv')
```

```python
#Displaying first 20 rows for reference
cp_df.head(20)
```

```python
#q1: Melt
q1 = cp_df[['Date','BUDAPEST']]
q1.melt(id_vars='BUDAPEST')
```

```python
#q2: Pivot 
q2 = cp_df[['Date','BUDAPEST','ZALA']]

q2.pivot(index='BUDAPEST', columns='Date')
#I ccould't think of a good way to use pivot with this dataset
```

```python
#q3: Aggregate
q3 = cp_df.aggregate(['sum','min','max'])
q3
```

```python
#q4: Iteration
q4 = cp_df

for index, row in q4.iterrows():
    print(row['BUDAPEST'])
    
```
